# 7. Removing Stop Words

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(text):
    new_text = []
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
# To remove emojis
import re
def remove_emojis(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"
                               u"\U0001F300-\U0001F5FF"
                               u"\U0001F680-\U0001F6FF"
                               u"\U0001F1E0-\U0001F1FF"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

 fater than lemmetisation, shows to user because it is faster
 #### Find out root/stem words, ex = walk, walking, walked, walks
 ## Use on Information Reterval Systems, ex- Google

```python
# NLTK , Stemmer Algorithm
# 1. Porter Stemmer(english)
# 2. Snow Boul Stemmer(other language)
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

# 11. Lemmatization
it is finding the original english word from dictionary thats why it is slower than stemming

```python
import nltk
from nltk.stem import WordNetLemmatizer
```

```python
wordnet_lemma = WordNetLemmatizer()
sentence = 'He was palying chess and eating at same time. He has a bad habit of swimming after playing long hours'
puncutations = "?;!.,:"
sentence_words = nltk.word_tokenize(sentence)
for word in sentence_words:
    if word in puncutations:
        sentence_words.remove(word)
        
sentence_words
print("{0:20}{1:20}".format("Word", "Lemma"))
for word in sentence_words:
    print ("{0:20}{1:20}".format(word,wordnet_lemma.lemmatize(word, pos='v'))) # POS- Parts of Speech
```
